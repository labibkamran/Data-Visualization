## Task E – Visual Gallery

This section builds a line of releases per year, a bar of top genres, a histogram of global sales, a regional pie, and boxplots for the best-selling platforms. These cover the basic trends.


```python
# Task E - 1: Line plot of games released per year
import pandas as pd
import matplotlib.pyplot as plt

# Convert Year to numeric and count releases per year
_years = pd.to_numeric(games_df['Year'], errors='coerce').dropna().astype(int)
releases_per_year = _years.value_counts().sort_index()

plt.figure(figsize=(10, 4))
releases_per_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.title('Games Released Per Year')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Task E - 2: Suitable chart of top 10 game genres
import matplotlib.pyplot as plt

genre_counts = games_df['Genre'].value_counts().head(10)
plt.figure(figsize=(8, 4))
genre_counts.plot(kind='bar', color='#4C72B0')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 10 Game Genres (by count)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Task E - 3: Histogram of Global_Sales distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
_games_global_sales = pd.to_numeric(games_df['Global_Sales'], errors='coerce').dropna()
plt.hist(_games_global_sales, bins=30, color='#55A868', edgecolor='white')
plt.xlabel('Global Sales (millions)')
plt.ylabel('Frequency')
plt.title('Distribution of Global Sales')
plt.tight_layout()
plt.show()
```

```python
# Task E - 4: Pie chart of regional sales percentages
import matplotlib.pyplot as plt

regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
regional_totals = games_df[regions].sum()
plt.figure(figsize=(6, 6))
plt.pie(regional_totals, labels=regions, autopct='%1.1f%%', startangle=90, colors=['#4C72B0', '#55A868', '#C44E52', '#8172B3'])
plt.title('Regional Sales Share')
plt.axis('equal')
plt.show()
```

```python
# Task E - 5: Box plot of sales by platform for top 5 platforms
import matplotlib.pyplot as plt

# Determine top 5 platforms by total Global_Sales
top_platforms = games_df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(5).index
subset = games_df[games_df['Platform'].isin(top_platforms)]

plt.figure(figsize=(10, 5))
# Create a list of arrays, one per platform, for boxplot
box_data = [subset[subset['Platform'] == p]['Global_Sales'].values for p in top_platforms]
plt.boxplot(box_data, labels=top_platforms)
plt.xlabel('Platform')
plt.ylabel('Global Sales (millions)')
plt.title('Global Sales Distribution by Platform (Top 5 Platforms)')
plt.tight_layout()
plt.show()
```

## Task F – GroupBy Highlights

I compute mean sales per platform, total sales per genre, yearly release counts, platform–genre averages, and the publishers with the highest average sales so I can cite numbers later.


```python
# Task F - 1: Group by Platform and calculate mean Global_Sales
platform_mean_sales = (games_df
    .groupby('Platform', as_index=False)['Global_Sales']
    .mean()
    .sort_values('Global_Sales', ascending=False)
)
print(platform_mean_sales.head())
```

```python
# Task F - 2: Group by Genre and find total sales for each genre
genre_total_sales = (games_df
    .groupby('Genre', as_index=False)['Global_Sales']
    .sum()
    .sort_values('Global_Sales', ascending=False)
)
print(genre_total_sales.head())
```

```python
# Task F - 3: Group by Year and count number of games
year_counts = (games_df
    .assign(Year=pd.to_numeric(games_df['Year'], errors='coerce').astype('Int64'))
    .dropna(subset=['Year'])
    .groupby('Year', as_index=False)['Name']
    .count()
    .rename(columns={'Name': 'Count'})
    .sort_values('Year')
)
print(year_counts.head())
```

```python
# Task F - 4: Group by Platform and Genre, calculate mean sales
platform_genre_mean = (games_df
    .groupby(['Platform', 'Genre'], as_index=False)['Global_Sales']
    .mean()
    .sort_values(['Platform', 'Global_Sales'], ascending=[True, False])
)
print(platform_genre_mean.head())
```

```python
# Task F - 5: Find the publisher with the highest average Global_Sales
publisher_avg = (games_df
    .groupby('Publisher', as_index=False)['Global_Sales']
    .mean()
    .sort_values('Global_Sales', ascending=False)
)
print(publisher_avg.head(10))
print("\nTop publisher by average Global_Sales:")
print(publisher_avg.iloc[0])
```

## Task G – Light Features

I add helper columns like decade, sales buckets (Low/Medium/High), uppercase names, total regional sales, and a performance tag. These make later comparisons easy.


```python
# Task G - 1: Create a decade column from Year
year_num = pd.to_numeric(games_df['Year'], errors='coerce')
games_df['Decade'] = (year_num // 10 * 10).astype('Int64')
print(games_df[['Year', 'Decade']].head())
```

```python
# Task G - 2: Categorize Global_Sales into Low/Medium/High
bins = [float('-inf'), 1.0, 5.0, float('inf')]
labels = ['Low', 'Medium', 'High']
games_df['Sales_Category'] = pd.cut(games_df['Global_Sales'], bins=bins, labels=labels)
print(games_df[['Global_Sales', 'Sales_Category']].head())
```

```python
# Task G - 3: Convert all game names to uppercase
games_df['Name_Upper'] = games_df['Name'].str.upper()
print(games_df[['Name', 'Name_Upper']].head())
```

```python
# Task G - 4: Create a new column for total regional sales
games_df['Total_Regional_Sales'] = games_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum(axis=1)
print(games_df[['Total_Regional_Sales', 'Global_Sales']].head())
```

```python
# Task G - 5: Apply a function to categorize games by sales performance
# Custom function using both Global_Sales and regional distribution as an example

def performance_category(row):
    g = row['Global_Sales']
    if g >= 10:
        return 'Blockbuster'
    elif g >= 5:
        return 'Hit'
    elif g >= 1:
        return 'Moderate'
    else:
        return 'Niche'

games_df['Performance'] = games_df.apply(performance_category, axis=1)
print(games_df[['Name', 'Global_Sales', 'Performance']].head())
```

```python
# Task H - 2: Top 10 best-selling games (by Global_Sales)
import pandas as pd

try:
    games_df
except NameError:
    games_df = pd.read_csv('vgsales.csv')

# Sort by Global_Sales descending and take top 10
top10 = games_df.nlargest(10, 'Global_Sales')[['Name', 'Platform', 'Year', 'Publisher', 'Global_Sales']]
print(top10)

# Optional: simple bar chart
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.barh(top10['Name'][::-1], top10['Global_Sales'][::-1], color='#4C72B0')
plt.title('Top 10 Best-Selling Games (Global Sales)')
plt.xlabel('Global Sales (millions)')
plt.ylabel('Game')
plt.tight_layout()
plt.show()
```

```python
# Task H - 3: Sales analysis by platform
import pandas as pd
import matplotlib.pyplot as plt

try:
    games_df
except NameError:
    games_df = pd.read_csv('vgsales.csv')

# Aggregate multiple statistics per platform
platform_sales = (games_df
    .groupby('Platform')
    .agg(
        Total_Global_Sales=('Global_Sales', 'sum'),
        Avg_Global_Sales=('Global_Sales', 'mean'),
        Median_Global_Sales=('Global_Sales', 'median'),
        Games_Count=('Name', 'count')
    )
    .sort_values('Total_Global_Sales', ascending=False)
)
print(platform_sales.head(10))

# Visualization: bar chart for total global sales by platform (top 10)
topN = 10
platform_top = platform_sales.head(topN)
plt.figure(figsize=(10, 5))
platform_top['Total_Global_Sales'].plot(kind='bar', color='#55A868')
plt.title(f'Top {topN} Platforms by Total Global Sales')
plt.xlabel('Platform')
plt.ylabel('Total Global Sales (millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization: scatter of avg sales vs number of games
plt.figure(figsize=(7, 5))
plt.scatter(platform_sales['Games_Count'], platform_sales['Avg_Global_Sales'], c='#4C72B0', alpha=0.7)
plt.title('Avg Global Sales vs. Number of Games (per Platform)')
plt.xlabel('Number of Games')
plt.ylabel('Average Global Sales (millions)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Task H - 4: Genre popularity over time
import pandas as pd
import matplotlib.pyplot as plt

try:
    games_df
except NameError:
    games_df = pd.read_csv('vgsales.csv')

# Prepare Year as numeric
_df = games_df.copy()
_df['Year'] = pd.to_numeric(_df['Year'], errors='coerce')
_df = _df.dropna(subset=['Year'])
_df['Year'] = _df['Year'].astype(int)

# Count number of releases per year per genre
genre_year_counts = (_df
    .groupby(['Year', 'Genre'])['Name']
    .count()
    .reset_index(name='Count')
)

# Pivot for plotting
pivot = genre_year_counts.pivot(index='Year', columns='Genre', values='Count').fillna(0)

plt.figure(figsize=(12, 6))
for col in pivot.columns:
    plt.plot(pivot.index, pivot[col], label=col, linewidth=1)
plt.title('Genre Popularity Over Time (Releases per Year)')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.legend(bbox_to_anchor=(1.04, 1), loc='upper left', ncol=1, fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
